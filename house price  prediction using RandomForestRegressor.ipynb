# import libiries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

# import dataset

```python
data = pd.read_csv('housing.csv')
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
X =data.drop(['median_house_value'] ,  axis=1)
y =data['median_house_value']
```

```python

X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.20)
```

```python
train_data = X_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize=(15,8))
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr() ,annot=True ,cmap=None)
```

```python
train_data.ocean_proximity.value_counts()
```

```python
train_data=train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'],axis=1) 
```

```python
train_data=train_data.drop(['ISLAND'],axis=1)
```

```python
train_data
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr() ,annot=True ,cmap=None)
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(data=train_data , x='latitude' , y='longitude' ,hue='median_house_value', palette='coolwarm')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler


X_train , y_train = train_data.drop(['median_house_value'] , axis=1) , train_data['median_house_value']
reg=LinearRegression()
scale = StandardScaler()
X_train_s = scale.fit_transform(X_train)

reg.fit(X_train_s , y_train)
```

```python
train_data
```

```python
test_data = X_test.join(y_test)
test_data=test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'],axis=1) 
```

```python
test_data= test_data.drop(['ISLAND'] , axis=1)
```

```python
X_test , y_test = test_data.drop(['median_house_value',] , axis=1) , test_data['median_house_value']
```

```python
test_data
```

```python
X_test_s = scale.transform(X_test)
```

```python
reg.score(X_test_s , y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor()
forest.fit(X_train_s , y_train)
```

```python
forest.score(X_test_s , y_test)
```

```python
from sklearn.model_selection import GridSearchCV
forest = RandomForestRegressor()
param_grid={
    "max_depth" :[1,2,3,4,5,6,8,10],
    "max_features" :['sqrt', 'log2'],
    "n_estimators" :[100,200,300]
    
}
```

```python
Grid_search = GridSearchCV(forest , param_grid , cv=5 , scoring="neg_mean_squared_error") 
Grid_search.fit(X_test_s , y_test)
```

```python
Grid_search.best_estimator_.score(X_test_s , y_test)
```
